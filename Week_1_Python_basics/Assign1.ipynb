```python
# Importing required libraries
import pandas as pd
import numpy as np
```

```python
# Loading the CSV file

df = pd.read_csv("D:\Celebal Assignments\Week1_assign\Combined_dataset.csv")

# Displaying first few rows
df.head()
```

```python
# Checking dataset shape
print("Shape of dataset:", df.shape)

# Column names
print("\nColumns:")
print(df.columns)

# Data types
print("\nData Types:")
print(df.dtypes)
```

```python
# First 5 rows
print("First 5 Records:")
display(df.head())

# Last 5 rows
print("Last 5 Records:")
display(df.tail())
```

```python
# Finding missing values in each column
print("Missing Values:")
print(df.isnull().sum())
```

```python
# Filling missing numerical values with mean

df['discount'] = df['discount'].fillna(df['discount'].mean())

# Filling missing text values with 'Not Available'

df['seller_name'] = df['seller_name'].fillna('Not Available')
df['seller_information'] = df['seller_information'].fillna('Not Available')
df['what_customers_said'] = df['what_customers_said'].fillna('Not Available')
df['videos'] = df['videos'].fillna('Not Available')
df['variations'] = df['variations'].fillna('Not Available')

print("Missing values after cleaning:")
print(df.isnull().sum())
df
```

```python
print(df['final_price'].head())
print(df['final_price'].dtype)
```

```python
# Remove quotes, currency symbol and commas

df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace('"', '', regex=False)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
)

# Convert to numeric
df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')

print(df['final_price'].head())
print(df['final_price'].dtype)
```

```python
# Using ratings_count as a proxy metric since quantity data is unavailable
#SINCE QUANTITY COLUMN IS NOT GIVEN *************************************** 
df['engagement_value'] = df['final_price'] * df['ratings_count']

df[['final_price', 'ratings_count', 'engagement_value']]
```

```python
# Saving the cleaned dataset as a new CSV file

df.to_csv("cleaned_dataset.csv", index=False)

print("Cleaned dataset saved successfully.")
```
